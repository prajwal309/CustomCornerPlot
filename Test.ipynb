Import the basic libraries such as numpy and also the CornerPlot

```python
#Load all the modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import CornerPlot


```

Create a basic corner plot for fitting plot 

```python
Parameters = ["m","b"]
NumPoints = 10000
mValues = np.random.normal(2.34,0.1, NumPoints)
bValues = np.random.normal(5.56, 0.2, NumPoints)

Data2Plot = np.vstack((mValues, bValues))

CornerPlot.CustomCornerPlot(Data2Plot,Parameters)
```

```python
Parameters = ["m","b"]
NumPoints = 10000

m1Values = np.random.normal(2.34,0.1, NumPoints)
b1Values = np.random.normal(5.56, 0.2, NumPoints)
Data1 = np.vstack((m1Values, b1Values))

m2Values = np.random.normal(4.67,0.1, NumPoints)
b2Values = np.random.normal(9.56, 0.2, NumPoints)
Data2 = np.vstack((m2Values, b2Values))


CornerPlot.MultipleCustomCornerPlot(Data1,Data2, Parameters)
```
